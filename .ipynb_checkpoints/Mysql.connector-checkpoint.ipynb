```python
FMT_QUERY = "Query {0}:\n" + "-"*8
FMT_HEADER = "{0:18s} {1:7s} {2:3s}"
FMT_ROW = "{0:18s} {1:^7s} {2:4.1f}"
```

```python
sql = """SELECT Name, CountryCode, Population FROM world.city WHERE CountryCode = %s ORDER BY Population DESC LIMIT 3
"""
```

```python
db = mysql.connector.connect(option_files="connector.cnf")
```

```python
cursor = db.cursor(prepared = True)
```

```python
count = 0
for country in ("USA", "IND"):
    count += 1
    print(FMT_QUERY.format(count))
    cursor.execute(sql, (country,))
    print("Statement: {0}".format(cursor.statement))
    print("Row count: {0}".format(cursor.rowcount))
    if (cursor.with_rows):
        cursor.fetchall()
    print(FMT_HEADER.format("City", "Country", "Pop"))
    city = cursor.fetchone()
    while(city):
        print(FMT_ROW.format(city[0], city[1], city[2]/1000000))
        city = cursor.fetchone()
    print("")
cursor.close()
db.close()
```
