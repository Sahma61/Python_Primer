```python
FMT_QUERY = "Query {0}:\n" + "-"*8
FMT_HEADER = "{0:18s} {1:7s} {2:3s}"
FMT_ROW = "{0:18s} {1:^7s} {2:4.1f}"
```

```python
sql = """SELECT Name, CountryCode, Population FROM world.city WHERE CountryCode = %s ORDER BY Population DESC LIMIT 3
"""
```

```python
db = mysql.connector.connect(option_files="connector.cnf")
```

```python
cursor = db.cursor(prepared = True)
```

```python
cursor.execute(sql, ('IND', ))
print(cursor.fetchall())
```

```python
count = 0
for country in ("USA", "IND"):
    count += 1
    print(FMT_QUERY.format(count))
    cursor.execute(sql, (country, ))
    if cursor.with_rows:
        print(cursor.fetchall())
    print("Statement: {0}".format(cursor.statement))
    print("Row count: {0}".format(cursor.rowcount))
    print(FMT_HEADER.format("City", "Country", "Pop"))
    cursor.execute(sql, (country, ))
    city = cursor.fetchone()
    while(city):
        print(FMT_ROW.format(city[0], city[1], city[2]/1000000))
        city = cursor.fetchone()
    print("")
cursor.close()
db.close()
```

```python
import mysql.connector
from datetime import datetime
from time import sleep
from mysql.connector.conversion import MySQLConverter

FMT_QUERY = "Query {0} - {1}:\n" + "-"*19
FMT_HEADER = "{0:18s} {1:7s} {2:3s}"
FMT_ROW = "{0:18s} {1:^7s} {2:4.1f}"

db = mysql.connector.connect(option_files="connector.cnf", use_pure = True)
converter = MySQLConverter(db.charset, True)

sql1 = """
SELECT Name,  CountryCode, Population FROM world.city WHERE CountryCode = 'USA' ORDER BY POPULATION DESC LIMIT 3
"""
sql2 = "DO SLEEP(3)"

sql3 = """
SELECT Name,  CountryCode, Population FROM world.city WHERE CountryCode = 'IND' ORDER BY POPULATION DESC LIMIT 3
"""

queries = [sql1, sql2, sql3]
results = db.cmd_query_iter(";".join(queries))
count = 0
for result in results:
    count += 1
    time = datetime.now().strftime("%H:%M:%S")
    print(FMT_QUERY.format(count, time))
    if 'columns' in result:
        print(FMT_HEADER.format("City", "Country", "Pop"))
        (city, eof) = db.get_row()
        while(city):
            values = converter.row_to_python(
            city, result['columns'])
            print(FMT_ROW.format(values[0], values[1], values[2]/1000000))
            (city, eof) = db.get_row()
    else:
        print("No result to print.")
    sleep(2)
    print("")
db.close()  
```

```python
import mysql.connector
from datetime import datetime
from time import sleep

FMT_QUERY = "Query {0} - {1}:\n" + "-"*19
FMT_HEADER = "{0:18s} {1:7s} {2:3s}"
FMT_ROW = "{0:18s} {1:^7s} {2:4.1f}"

db = mysql.connector.connect(option_files="connector.cnf", use_pure = True)
cursor = db.cursor()

sql1 = """
SELECT Name,  CountryCode, Population FROM world.city WHERE CountryCode = 'USA' ORDER BY POPULATION DESC LIMIT 3
"""
sql2 = "DO SLEEP(3)"

sql3 = """
SELECT Name,  CountryCode, Population FROM world.city WHERE CountryCode = 'IND' ORDER BY POPULATION DESC LIMIT 3
"""

queries = [sql1, sql2, sql3]
results = cursor.execute(";".join(queries), multi = True)
count = 0
for result in results:
    count += 1
    time = datetime.now().strftime("%H:%M:%S")
    print(FMT_QUERY.format(count, time))
    if result.with_rows:
        print(FMT_HEADER.format("City", "Country", "Pop"))
        city = cursor.fetchone()
        while(city):
            print(FMT_ROW.format(city[0], city[1], city[2]/1000000))
            city = cursor.fetchone()
    else:
        print("No result to print.")
    sleep(2)
    print("")
db.close()  
```
